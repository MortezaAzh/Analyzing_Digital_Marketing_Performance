```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sample_data = pd.read_excel('data.xlsx',sheet_name='Data1-Daily channel online')
sample_data['ConversionRate'] = sample_data['ConversionRate'].astype(dtype = 'float64')
sample_data['Visitors'] = sample_data['Visitors'].astype(dtype = 'int64')
sample_data['Visits'] = sample_data['Visits'].astype(dtype = 'int64')
sample_data['Pageviews'] = sample_data['Pageviews'].astype(dtype = 'int64')
sample_data['basketstarts'] = sample_data['basketstarts'].astype(dtype = 'int64')
sample_data['bounces'] = sample_data['bounces'].astype(dtype = 'int64')
sample_data['Orders'] = sample_data['Orders'].astype(dtype = 'int64')
```

```python
sample_data
```

```python
email = sample_data[sample_data.Channel == 'email']
Natural_search = sample_data[sample_data.Channel == 'Natural search']
Paid_Search = sample_data[sample_data.Channel == 'Paid Search']
Direct = sample_data[sample_data.Channel == 'Direct']
Total = sample_data[sample_data.Channel == 'Total']
```

```python
plt.plot(email.Date,email.Orders)
plt.plot(Natural_search.Date,Natural_search.Orders)
plt.plot(Paid_Search.Date,Paid_Search.Orders)
plt.plot(Direct.Date,Direct.Orders)
plt.plot(Total.Date,Total.Orders)
plt.legend(['email','Natural_search','Paid_Search','Direct','Total'],prop={'size': 8},loc = 'upper left')
plt.xlabel('Date')
plt.ylabel('Orders')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.plot(email.Date,email.Visits)
plt.plot(Natural_search.Date,Natural_search.Visits)
plt.plot(Paid_Search.Date,Paid_Search.Visits)
plt.plot(Direct.Date,Direct.Visits)
plt.plot(Total.Date,Total.Visits)
plt.legend(['email','Natural_search','Paid_Search','Direct','Total'],prop={'size': 8},loc = 'upper left')
plt.xlabel('Date')
plt.ylabel('Visits')
plt.gcf().autofmt_xdate()
plt.show()
```
